# Sentiment Analysis


*   	Methodology: First, we will use NLTK library to do a sentiment analysis and classify all reviews into the following categories: positive, negative and neutral based on the sentiment score received.






```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
def sentiment_analysis(combined_df):
    sid = SentimentIntensityAnalyzer()
    sentiment_score = sid.polarity_scores(combined_df)
    return sentiment_score['compound']
combined_df['Sentiment_Score'] = combined_df['content'].apply(sentiment_analysis)
```

```python
combined_df['Sentiment_Score'] = combined_df['content'].apply(sentiment_analysis)
```

```python
def classify_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
combined_df['Sentiment'] = combined_df['Sentiment_Score'].apply(classify_sentiment)
```

```python
combined_df.head()
```

```python
# Filter for positive sentiment
positive_df = combined_df[combined_df['Sentiment'] == 'Positive']

# Count the occurrences of each product ID
positive_counts = positive_df['product_id'].value_counts()

# Create the bar chart
positive_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Count of Positive Sentiments by Product ID')
plt.xlabel('Product ID')
plt.ylabel('Count of Positive Sentiments')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

We see Bose has the highest number of positive reviews, which seconds our findings of Bose having the highest average ratings as well.

*   SWOT Analysis



**Price Sensitivity:**

* The headphone market is highly competitive, with consumers expecting a strong feature set at various price points.
Premium brands like Apple AirPods Max and Bose QuietComfort 45 face pressure to justify their high prices.
Sound Quality:

* While sound quality is generally praised, specific criticisms include brightness, muddy bass, and lack of customization options.
There's an opportunity for brands to differentiate themselves by addressing these sound-related issues.
Comfort and Fit:

* Ear cup size, clamping force, and overall comfort are critical factors.
Brands should prioritize ergonomic designs to enhance user experience.
Noise Cancellation (ANC):

* ANC is a key selling point, but performance varies widely.
There's potential for brands to differentiate by offering superior ANC capabilities.
Build Quality and Durability:

* Consumers expect durable products. Plastic construction, while cost-effective, can negatively impact brand perception.
Investing in higher-quality materials can improve customer satisfaction.
Additional Features:

* Features like auto-switching, app integration, and voice assistant compatibility are becoming increasingly important.
Brands should focus on providing seamless user experiences through these features.
\\

**Competitive Landscape**

* Apple AirPods Max: While offering premium sound quality, the high price, weight, and lack of auto-pause are significant drawbacks.
Bose QuietComfort 45: Despite a strong brand reputation, the older model faces competition in terms of price, battery life, and app functionality.
* Sony ULT Wear: Issues with build quality, comfort, and mic quality hinder its competitiveness.
* Beats Studio Pro: Strong sound quality but faces competition in terms of noise cancellation, build quality, and price.
* Beyerdynamic Lagoon ANC: Struggles with fundamental aspects like controls, pairing, and ANC, making it a less attractive option.
* JBL Live 660NC: Offers a decent balance of features but faces competition in terms of ear cup size, ANC performance, and app functionality.

**Opportunities for Differentiation**
* Enhanced ANC: Develop advanced noise cancellation technology to outperform competitors.
* Improved Build Quality: Invest in durable materials and construction to build trust with consumers.
* Optimized Comfort: Prioritize ergonomic design and user testing to ensure optimal comfort.
* Innovative Features: Introduce unique features like advanced audio customization or health monitoring.
* Competitive Pricing: Offer compelling value propositions by balancing features and price.
* Exceptional Customer Service: Build customer loyalty through excellent support and after-sales service.

# SWOT Analysis
**Strengths**
1. Sound Quality: While not universally praised, sound quality remains a strong
point for many models, with some excelling in specific areas like bass or clarity.
2. Noise Cancellation (ANC): Although effectiveness varies, ANC is a feature that is increasingly valued by consumers and is present in most models.
Feature Set: Some models offer impressive feature sets, including app integration, automatic switching, and advanced customization options.
**Weaknesses**
1. Price: High prices are a common complaint, especially considering that not all models deliver on the premium experience.
Build Quality: Consistency issues with build quality are evident, with some models feeling cheap or experiencing durability problems.
2. Comfort: Fit and comfort are subjective, but tight clamping force and small ear cups are recurring issues.
3. Battery Life: While generally sufficient, some models fall short of competitors in terms of battery life.
4. App Integration: Inconsistent app experiences, with some offering limited functionality or requiring multiple apps for different devices.
**Opportunities**
1. Differentiation: Brands can differentiate themselves by focusing on specific pain points, such as improving build quality, enhancing ANC, or offering superior comfort.
2. Customization: Providing greater customization options, such as adjustable EQ or personalized sound profiles, can enhance user experience.
3. Battery Life: Extending battery life can be a significant selling point, especially for frequent travelers.
4. Sustainability: Emphasizing eco-friendly materials and manufacturing processes can appeal to environmentally conscious consumers.
**Threats**
1. Intense Competition: The headphone market is highly competitive, with new models and technologies emerging rapidly.
2. Consumer Expectations: As technology advances, consumer expectations for features and performance will continue to rise.
3. Economic Factors: Economic downturns can impact consumer spending on premium products like headphones.
